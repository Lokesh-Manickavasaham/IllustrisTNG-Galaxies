```python
import illustris_python as il
import numpy as np
import matplotlib.pyplot as plt
```

```python
basePath = '../sims.TNG/TNG100-1/output/'
```

0a. Choose one or two entries in the group catalog which sound interesting, and plot their distribution(s).

```python
fields_a = ['GroupGasMetallicity', 'GroupStarMetallicity']
```

```python
halos = il.groupcat.loadHalos(basePath, 99, fields=fields_a)
```

```python
halos.keys()
```

```python
halos['GroupGasMetallicity'].shape, halos['GroupStarMetallicity'].shape, halos['count']
```

```python
np.count_nonzero(halos['GroupGasMetallicity']), np.count_nonzero(halos['GroupStarMetallicity'])
```

```python
fig, ax = plt.subplots()

ax.plot(halos['GroupGasMetallicity'], halos['GroupStarMetallicity'], '.', markersize=0.5)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Mass-weighted average metallicity (gas)')
ax.set_ylabel('Mass-weighted average metallicity (star)')

plt.show()
```
