```python
import illustris_python as il
import numpy as np
import matplotlib.pyplot as plt
```

```python
basePath = '../sims.TNG/TNG100-1/output/'
```

0b. Choose one or two entries in the snapshot which sound interesting, and plot their distribution(s).

```python
fields_b = ['GFM_StellarFormationTime']
```

```python
snaps = il.snapshot.loadSubset(basePath, 99, 4, fields=fields_b) # given as scalefactor
```

```python
snaps.shape
```

```python
mean = np.mean(snaps)
```

```python
fig, ax = plt.subplots()

ax.hist(snaps, bins = 500, label='Distribution of Star Formation Time')
ax.set_xlabel('Star Formation Time (scalefactor)')
ax.set_ylabel('No. of stars')
ax.axvline(x = mean, color= 'black', ls='--', label='Mean Star Formation Time')
ax.legend(fontsize='x-small')

plt.tight_layout()
plt.show()
```

An interesting thing to note is that the negative values of star formation time represent that these are for wind gas cells and positive values are for real stars.
